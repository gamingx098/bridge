## 📉 Volatility Regime Analysis (GARCH)

```python
# Function to analyze volatility
def garch_vol_analysis(ticker):
    # Download data
    df = yf.download(ticker, start='2018-01-01', end='2024-01-01')
    
    # Calculate returns
    df['Returns'] = df['Adj Close'].pct_change()
    df['Log_Returns'] = np.log(df['Adj Close'] / df['Adj Close'].shift(1))
    
    # GARCH model (using ARCH package)
    model = arch_model(df['Returns'].dropna(), vol='Garch', p=1, o=0, q=1)
    fitted_model = model.fit(disp='off')
    
    print(f"GARCH Analysis for {ticker}")
    print(fitted_model.summary())
    return df

# Analyze a sample stock (BBCA.JK)
garch_results = garch_vol_analysis('BBCA.JK')
```

## 📊 Correlation Matrix: Commodities vs JKSE

```python
# Create correlation matrix with commodities

# Sample commodity data (in practice, you'll use real-time or historical data)
commodities = ['CL.JK', 'GC.JK']  # Crude Oil and Gold

# Download market data
try:
    idx_data = yf.download('^JKSE', start='2023-01-01', end='2024-01-01')
    
    # Create DataFrame with returns
    df_returns = pd.DataFrame()
    df_returns['JKSE'] = idx_data['Adj Close'].pct_change().dropna()
    
    for commodity in commodities:
        try:
            commod_data = yf.download(commodity, start='2023-01-01', end='2024-01-01')
            df_returns[commodity] = commod_data['Adj Close'].pct_change().dropna()
        except:
            print(f"Could not fetch data for {commodity}")

    # Calculate correlation matrix
    corr_matrix = df_returns.corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Matrix: Commodities vs JKSE')
    plt.show()

except Exception as e:
    print(f"Error fetching data: {e}")
```
